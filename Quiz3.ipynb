```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('image.jpg')

# Plot the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original image')
plt.show()
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Plot the grayscale image
plt.imshow(gray, cmap='gray')
plt.title('Grayscale image')
plt.show()
```

```python
# Binarize the image using Otsu's thresholding
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Plot the binarized image
plt.imshow(thresh, cmap='gray')
plt.title('Binarized image')
plt.show()
```

```python
# Find edges in the image using Canny edge detection
edges = cv2.Canny(thresh, 100, 200)

# Plot the edge map
plt.imshow(edges, cmap='gray')
plt.title('Edge map')
plt.show()
```

```python
# Perform Hough transform to detect circles
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, 1, 20, param1=90, param2=77, minRadius=0, maxRadius=0)
```

```python
# Draw the circles on the image
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # Draw the outer circle
        cv2.circle(image, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # Draw the center of the circle
        cv2.circle(image, (i[0], i[1]), 2, (0, 0, 255), 3)
```

```python
# Plot the image with the detected Pupil
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Detected Pupil')
plt.show()
```
